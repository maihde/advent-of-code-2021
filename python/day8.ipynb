```python
with open("../dat/day8.txt") as f:
    notes = f.readlines()

data = []
for nn in notes:
    input_values, output_values = nn.split("|")
    input_values = [ xx.strip() for xx in input_values.split() ]
    output_values = [ xx.strip() for xx in output_values.split() ]
    data.append((input_values, output_values))

# 0 uses 6
# 1 uses 2
# 2 uses 5
# 3 uses 5
# 4 uses 4
# 5 uses 5
# 6 uses 6
# 7 uses 3
# 8 uses 7
# 9 uses 6
cnt = 0
for input_values, output_values in data:
    for vv in output_values:
        vv = "".join(sorted(vv))
        if len(vv) in (2,4,3,7):
            cnt += 1
print(cnt)
```

```python
def solve(input_values, output_values):
    # Seed using known lengths
    segments = {} # wire => segment
    lengths = {} # length => set()
    digits = {} # digit => set()

    for x in input_values:
        lengths.setdefault(len(x), set()).add("".join(sorted(x)))
    
    # Certian digits can solely be identified based on their length
    if len(lengths.get(2, set())) == 1:
        digits[1] = lengths[2].pop()
    if len(lengths.get(3, set())) == 1:
        digits[7] = lengths[3].pop()
    if len(lengths.get(4, set())) == 1:
        digits[4] = lengths[4].pop()
    if len(lengths.get(7, set())) == 1:
        digits[8] = lengths[7].pop()

    if 7 not in digits and 1 not in digits:
        raise ValueError('cannot solve')

    # 7 and 1 only differ by one segment
    tmp = set(digits[7]) - set(digits[1])
    segments['a'] = "".join(tmp)

    # Of length 5 sequences, after removing digit 4
    # only digit 2 will create three segements
    # 2 - 4 = aeg
    # 3 - 4 = ag
    # 5 - 4 = ag
    for vv in list(lengths[5]):
        tmp = set(vv) - set(digits[4])
        if len(tmp) == 3:
            digits[2] = vv
            lengths[5].remove(vv)
        elif len(tmp) == 2:
            tmp = tmp - set(segments['a'])
            tmp = "".join(tmp)
            if segments.setdefault('g', tmp) != tmp:
                raise ValueError('conflicting solve for g')

    if digits.get(2) is None:    
        raise ValueError('cannot solve')
    if segments.get('g') is None:
        raise ValueError('cannot solve')

    # 4 + 'a' + 'g' = 9
    tmp = set(digits[4]) | set(segments['a']) | set(segments['g'])
    tmp = "".join(sorted(tmp))
    for vv in list(lengths[6]):
        if tmp == vv:
            if digits.get(9) is None:
                digits[9] = vv
                lengths[6].remove(vv)
            else:
                raise ValueError("inconsistent solve for digit 9")
    
    if digits.get(9) is None:
        raise ValueError('cannot solve')

    # 8 - 9 = e
    tmp = set(digits[8]) - set(digits[9])
    tmp = "".join(tmp)
    segments['e'] = tmp

    # if you add segment 'e' to the remaining five lengths
    # 5 + 'e' will be in a length 6 which equals 6
    # 3 + 'e' will not
    tmp = set(vv) | set(segments['e'])
    tmp = "".join(sorted(tmp))

    if len(lengths[5]) != 2:
        raise ValueError('inconsistent solve')

    xx = lengths[5].pop()
    yy = lengths[5].pop()

    tmp_xx = set(xx) | set(segments['e'])
    tmp_xx = "".join(sorted(tmp_xx))

    tmp_yy = set(yy) | set(segments['e'])
    tmp_yy = "".join(sorted(tmp_yy))

    if tmp_xx in lengths[6]:
        digits[6] = tmp_xx
        digits[5] = xx
        digits[3] = yy
        lengths[6].remove(tmp_xx)
    elif tmp_yy in lengths[6]:
        digits[6] = tmp_yy
        digits[5] = yy
        digits[3] = xx
        lengths[6].remove(tmp_yy)
    else:
        raise ValueError('inconsistent solve')
    
    if len(lengths[6]) != 1:
        raise ValueError('inconsistent solve')
    digits[0] = lengths[6].pop()

    ans = []
    for vv in output_values:
        tmp = "".join(sorted(vv))
        for xx, yy in digits.items():
            if tmp == yy:
                ans.append(str(xx))

    return int("".join(ans))    
```

```python
sum = 0
for input_values, output_values in data:
    ans = solve(input_values, output_values)
    sum += ans
print(sum)
```
